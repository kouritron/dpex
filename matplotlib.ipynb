# ==================== plot

```python
x = np.arange(0, 5, 0.01)
y = np.sin(2*x)
plt.plot(x, y)
```

```python
# (x, y, fmt, **kwargs)
# x and y would have to be "array like" or scalar
# fmt: shorthand versions of some the things you can control with kwargs. 
# fmt == "ro" means red circles  ***  'r--' means red dashed line
plt.plot([1, 2, 3], [1, 2, 3], 'go-', label='line 1', lw=3, ls='-')
plt.plot([1, 2, 3], [1, 4, 9], 'rs', label='line 2')

# -----
x = np.arange(0.1, 3, 0.01)
# y = np.log((2+np.sin(10*x))*5)
y = np.sin((np.log(x))*8)
kw3 = {
    'label': 'sin(log(x)*8)',
    # visible: False,
    # visible: True,
}
plt.plot(x, y, 'b', **kw3)

plt.legend()
```

# ==================== pass to PIL

```python
# pass it to PIL and then you can use that to save/show ....
# again it seems that plt.show() is best avoided.

from matplotlib.backends.backend_agg import FigureCanvasAgg

x = np.arange(0.1, 5, 0.01)
plt.plot(x, np.sin(3*x), label='sin(3x)')
plt.legend()

agg = plt.get_current_fig_manager().canvas.switch_backends(FigureCanvasAgg)
agg.draw()
buf, size = agg.print_to_buffer()

im = Image.frombytes("RGBA", size, buf)

# Uncomment to show using external tool.
# im.show()
```
